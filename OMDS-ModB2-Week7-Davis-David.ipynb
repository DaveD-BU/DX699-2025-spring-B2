# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

````python
import numpy as np
import seaborn as sns
import pandas as pd
````

````python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
````

````python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
````

Here are a few possibilities to use preattentive attributes to draw attention to the last three months of data:

- Color: Use a distinct color for the last three months of data that contrasts with the rest of the dataset. For example, if your chart uses shades of blue for the rest of the data, you could use a bright red or green for the last three months.

- Size: Increase the size of the data points or bars representing the last three months. Larger elements naturally draw more attention.

- Position: Place the last three months of data in a prominent position within the chart. For instance, starting the chart from the most recent data and working backwards can make the last three months appear first.

- Shape: Use a different shape for the last three months of data points. If the rest of your data points are circular, using squares or triangles for the last three months will make them stand out.

- Orientation: Tilt or angle the last three months' data points differently from the rest. If all other data points are upright, angling these differently can attract attention.

````python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import TextArea, AnnotationBbox

# Data
concerns = [
    "Engine power is less than expected",
    "Tires make excessive noise while driving",
    "Engine makes abnormal/excessive noise",
    "Seat material concerns",
    "Excessive wind noise",
    "Hesitation or delay when shifting",
    "Bluetooth system has poor sound quality",
    "Steering system/wheel has too much play",
    "Bluetooth system is difficult to use",
    "Front seat audio/entertainment/navigation controls"
]
values = [12.9, 12.3, 11.6, 11.6, 11.0, 10.3, 10.0, 8.8, 8.6, 8.2]
colors = ['#D98CB3', '#d62728', '#d62728', '#D98CB3', '#d62728', '#D98CB3', '#D98CB3', '#B0B8B8', '#B0B8B8', '#B0B8B8']
comments = [
    "Comments indicate that noisy tire issues are most apparent in the rain.",
    "Complaints about engine noise commonly cited after the car had not been driven for a while.",
    "Excessive wind noise is noted primarily in freeway driving at high speeds."
]

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(concerns, values, color=colors)

# Add text annotations
for bar in bars:
    width = bar.get_width()
    ax.text(width - 0.5, bar.get_y() + bar.get_height() / 2, f'{width}', ha='center', va='center', color='white')

# Add comments
ax.text(13, 1.25, comments[0], fontsize=14, color='black')
ax.text(13, 2.25, comments[1], fontsize=14, color='black')
ax.text(13, 4.25, comments[2], fontsize=14, color='black')


# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_xticks([]) 

plt.text(-7, -1, 'Top 10 design concerns', fontsize=20, ha='left')
plt.text(-7, -3, 'Of the top design concerns three are noise-related, ', fontsize=15, ha='left', color='#d62728')
plt.text( 0, -0.65, 'concerns per 1,000', fontsize=14, ha='left', color='#B0B8B8')

plt.gca().invert_yaxis()
plt.show()
````
