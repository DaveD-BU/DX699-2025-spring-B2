```python
import numpy as np
import pandas as pd

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
```

```python
df_cyber_features = df_cyber.drop(columns=['Source IP Address','Destination IP Address', 'Payload Data', 'User Information', 'Device Information', 'Proxy Information'])
```

```python
categorical_features = df_cyber_features.select_dtypes(exclude=['number']).columns.tolist()
numeric_features     = df_cyber_features.select_dtypes(include=['number']).columns.tolist()

# Print results if you want|
print("Categorical Features:", categorical_features)
print("Numeric Features:", numeric_features)
```

```python
# from sklearn.preprocessing   import OrdinalEncoder

# ordinal_encoder = OrdinalEncoder()
# df_cyber_features[categorical_features] = ordinal_encoder.fit_transform(df_cyber_features[categorical_features])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=len(df_cyber_features[numeric_features].columns)//4 + 1, ncols=4, figsize=(20, 20))
axes = axes.flatten()

for i, feature in enumerate(df_cyber_features[numeric_features].columns):
    sns.histplot(df_cyber_features[feature], kde=True, ax=axes[i])
    axes[i].set_title(feature)

plt.tight_layout()
plt.show()
```
