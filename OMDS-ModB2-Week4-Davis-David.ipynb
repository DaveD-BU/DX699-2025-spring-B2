Histograms and boxplots are potent tools for visualizing data, but they each have their strengths and weaknesses. Here's a comparison to help you decide which one to use:

### Histograms

**Pros:**

- **Detailed Distribution:** Histograms provide a detailed view of the data distribution, showing the frequency of data points within specified bins.
- **Pattern Recognition:** Useful for identifying patterns, such as normal distribution, skewness, and modality (uni-modal, bi-modal, etc.).
- **Easy to Understand:** Intuitive and easy to understand for audiences with varying levels of statistical knowledge.

**Cons:**

- **Bin Size Sensitivity:** The appearance of the histogram can change significantly with different bin sizes. Choosing the right bin width is crucial.
- **Less Precise:** Can be less precise for small data sets as the distribution might look different depending on the chosen bin size.
- **Comparisons:** Comparing distributions can be difficult if the histograms are not plotted together on the same axes.

### Boxplots

**Pros:**

- **Summary Statistics:** A concise summary of the data distribution, including the median, quartiles, and potential outliers.
- **Outlier Detection:** Excellent for identifying outliers quickly.
- **Comparison:** Ideal for comparing distributions across different groups or categories, as multiple boxplots can be displayed side by side.

**Cons:**

- **Less Detail:** Does not provide as detailed a view of the data distribution as histograms. You can see the range, central tendency, and variability but not the exact shape of the distribution.
- **Interpretation:** May be less intuitive for audiences unfamiliar with boxplots or summary statistics.

### When to Use Each

- **Histograms:** Use histograms to get a detailed view of the data distribution and to identify patterns or specific distribution shapes.
- **Boxplots:** Use boxplots when you need a quick summary of the data distribution, want to compare distributions across different groups, or need to detect outliers.

Combining histograms and boxplots can often provide a comprehensive understanding of your data. For example, you might use a histogram to get a detailed view of the data distribution and a boxplot to summarize it and compare it across different groups.

```python
def clean_data(df):
    # Derive column 'date' from column: 'Timestamp'
    df.insert(1, 'date', df.apply(lambda row : "", axis=1))
    df['Timestamp'] = pd.to_datetime(df['Timestamp'])
    df['date'] = df['Timestamp'].dt.date
    # Drop columns: 'Timestamp', 'Source IP Address' and 4 other columns
    df = df.drop(columns=['Timestamp', 'Source IP Address', 'Destination IP Address', 'Payload Data', 'User Information', 'Device Information'])
    return df

def one_hot_encode(df):
    # One-hot encode columns: 'Protocol', 'Packet Type' and 7 other columns
    for column in ['Protocol', 'Packet Type', 'Traffic Type', 'Attack Type', 'Attack Signature', 'Action Taken', 'Severity Level', 'Network Segment', 'Log Source']:
        insert_loc = df.columns.get_loc(column)
        df = pd.concat([df.iloc[:,:insert_loc], pd.get_dummies(df.loc[:, [column]], drop_first=False).astype(int), df.iloc[:,insert_loc+1:]], axis=1)
    return df

# Loaded variable 'df' from URI: c:\repos\BU MSDS Repo\DX699\DX699-2025-spring-B2\cybersecurity_attacks.csv
df_cyber_preencoding = clean_data(df_cyber.copy())

df_cyber_clean = one_hot_encode(df_cyber_preencoding.copy())
df_cyber_clean.head()
```

```python
categorical_features = df_cyber_clean.select_dtypes(exclude=['number']).columns.tolist()
numeric_features     = df_cyber_clean.select_dtypes(include=['number']).columns.tolist()

# Print results if you want|
print("Categorical Features:", categorical_features)
print("Numeric Features:", numeric_features)
```

```python
labels = 'Supplier A', 'Supplier B', 'Supplier C', 'Supplier D'
sizes = [34, 31, 26, 9]
colors = ['#00008B', '#0000FF', '#ADD8E6', '#E0FFFF']
explode = (0.1, 0.1, 0.1, 0.1)  

# Plot
fig, ax = plt.subplots(figsize=(10, 5))

wedges, texts, autotexts = ax.pie(sizes, explode=explode, labels=None, colors=colors, autopct='%1.1f%%',
                                  shadow=True, startangle=140, textprops={'color': 'white'})

for autotext, color in zip(autotexts, colors):
    autotext.set_color('white' if color in ['#00008B', '#0000FF'] else '#0000FF')

ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax.legend(labels, loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_title('Supplier Market Share', loc='left', fontsize=16)

plt.show()
```
