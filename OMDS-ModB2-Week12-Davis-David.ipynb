# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

The disadvantage of the first graph is that it is difficult to see the trend over time.  The advantage of the second graph is that it is easy to see the trend over time.  In the first chart the scale of the y-axis is not the same for each day, so it is difficult to compare the number of visitors on different days.  In the second chart, the scale of the y-axis is the same for each day, so it is easy to compare the number of visitors on different days.  The disadvantage of both graphs is that it is difficult to see the individual data points. This can be fixed by adding an indicator at each dey.

```python
import matplotlib.pyplot as plt

# Years shown in the chart
years = list(range(2007, 2015))

# Approximate price data for each product line based on the chart
price_A = [None, 390, 410, 415, 385, 300, 275, 260]      # Product A
price_B = [None, 370, 400, 405, 375, 290, 270, 255]      # Product B
price_C = [None, None, None, 110, 180, 200, 230, 175]   # Product C
price_D = [None, None, None, None, 160, 260, 220, 210]  # Product D
price_E = [None, None, None, None, None,None, 100, 205] # Product E
avg_price = [None] * 7+ [223]                          # Only shown for 2014

# Create figure and axis
fig, ax = plt.subplots(figsize=(9, 6))

# Plot each product line
ax.plot(years, price_A, label='A', color='gray')
ax.plot(years, price_B, label='B', color='gray')
ax.plot(years, price_C, label='C', color='gray')
ax.plot(years, price_D, label='D', color='gray')
ax.plot(years, price_E, label='E', color='gray')

# Plot average price point for 2014
ax.plot(years, avg_price, 'o', color='blue', label='AVG')

# Move the point slightly to the left and add a label
if avg_price[-1] is not None:
    ax.text(years[-1] - 0.3, avg_price[-1], 'AVG', color='blue', fontsize=10, va='center')

# Highlight the recommended range ($150 - $200)
ax.axhspan(150, 200, color='lightblue', alpha=0.5)
ax.text(2007.1, 175, 'Recommended range', color='blue', fontsize=10, va='center')

#remove the top and right boarder lines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set axis limits and ticks
ax.set_ylim(0, 500)
# Extend the x-axis by 1/3 of the series
extra_space = (years[-1] - years[0]) / 99
ax.set_xlim(years[0], years[-1] + extra_space)
ax.set_xticks(years)
ax.set_xticklabels([str(year) if year != 2007 else '' for year in years])  
ax.set_ylabel('Average price')
ax.yaxis.set_label_coords(-0.08, 0.92)
ax.set_xlabel('Year')
ax.xaxis.set_label_coords(0.14, -0.06)

# Label the left end of each line with the product letter
for price, label in zip([price_A, price_B, price_C, price_D, price_E], ['A', 'B', 'C', 'D', 'E']):
    for i, value in enumerate(price):
        if value is not None:  # Find the first non-None value
            ax.text(years[i] - 0.15, value, label, color='gray', fontsize=11, va='center')
            break

# Set custom y-ticks and labels
y_ticks = [0, 150, 200, 300, 400, 500]
y_labels = ['$0', '$150', '$200', '$300', '$400', '$500']
ax.set_yticks(y_ticks)
ax.set_yticklabels(y_labels)

# Highlight 150 and 200 in light blue
# Increase the line width for all plot lines
for line in ax.get_lines():
    line.set_linewidth(2.5)

# Highlight 150 and 200 in light blue
for tick, label in zip(ax.get_yticks(), ax.get_yticklabels()):
    if tick in [150, 200]:
        label.set_color('blue')
        label.set_weight('bold')

top_label = ("To be competitive, we recommend introducing our product below the $223 average price point in the $150$200 range")
ax.text(2007 - .7, 555, top_label, fontsize=12, color='blue', ha='left', va='bottom', wrap=True)

title_label =("Retail price over time" )
ax.text(2007 - .7, 520, title_label, fontsize=14, ha='left', va='bottom', wrap=True)

# Display the plot
plt.tight_layout()
plt.show()
```
